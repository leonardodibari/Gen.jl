```julia
function compare_seq(a::Array{Int8,1}, b::Array{Int8,1})
    if a == b
        return 0
    else
        return findall(a .!= b)[1]
    end
end

function find_positions(msa)
    
    return [compare_seq(msa[j-1,:], msa[j,:]) for j in 2:length(msa[:,1])]
        
end


function mut_effect_WTtoB_one_seq(h::Array{Float64,2}, J::Array{Float64,4}, 
        seq, mut_s, b)
    
    seq_pre = deepcopy(seq)
    seq_post = deepcopy(seq)
    
    seq_post[mut_s] = b
    
    return delta_energy(h,J, seq_post, seq_pre)
    
end


function mut_effect_WTtoA_one_seq(h::Array{Float64,2}, J::Array{Float64,4}, 
        seq, mut_s, a)
    
    seq_pre = deepcopy(seq)
    seq_post = deepcopy(seq)
    
    seq_post[mut_s] = a
    
    return delta_energy(h,J, seq_post, seq_pre)
    
end

   
function cont_entr_over_time(h::Array{Float64,2}, J::Array{Float64,4}, 
        seq_chain::Array{Int8,2}, mut_t, mut_s, a, b, window)
    
    res = Array{Float64}(undef,2*window+1)
    idx = 0
    for i in mut_t-window:mut_t
        idx+=1
        
        if seq_chain[i,:] == seq_chain[i-1,:] 
            #println("no mut")
            if idx !=1
                res[idx] = res[idx-1]
            end
        else        
            res[idx] = mut_effect_WTtoB_one_seq(h, J, seq_chain[i,:], mut_s, b)
        end
    end

    for i in mut_t+1:mut_t+window
        idx+=1
        
        if seq_chain[i,:] == seq_chain[i-1,:] 
            #println("no mut")
            if idx !=1
                res[idx] = res[idx-1]
            end
        else        
            res[idx] = mut_effect_WTtoA_one_seq(h, J, seq_chain[i,:], mut_s, a)
        end
    end

    return res
end

function cont_entr_setofmuts(h::Array{Float64,2}, J::Array{Float64,4}, 
        seq_chain::Array{Int8,2}, mut_times::Array{Int64,1}, 
        mut_sites::Array{Int64,1}, a, b, n_muts, start_t,window)
    res = Array{Float64}(undef, n_muts, 2*window+1)
    for i in 1:n_muts
        res[i,:] = cont_entr_over_time(h, J, 
            seq_chain, mut_times[start_t+i], mut_sites[start_t+i], 
            a[start_t+i], b[start_t+i], window)
    end
    return res
end

function cont_entr_setofmuts_eachstep(h::Array{Float64,2}, J::Array{Float64,4}, 
        seq_chain::Array{Int8,2}, mut_times::Array{Int64,1}, 
        mut_sites::Array{Int64,1}, a, b, window, mut_range)
    res = Array{Float64}(undef, length(mut_range), 2*window+1)
    idx = 0
    for i in mut_range
        idx +=1
        res[idx,:] = cont_entr_over_time(h, J, seq_chain, mut_times[i], mut_sites[i], 
            a[i], b[i], window)
    end
    return res
end

function get_entropy(f)
    N=length(f[1,:])
    entr = zeros(Float64, N)
    for i in 1:N
        for a in 1:20
            if(f[a,i]>0)
                entr[i]-=f[a,i]*log(f[a,i])
            end
        end
    end
    
    return entr
end

function cont_dep_entr_without_deg(background, h, J)
    
    T = 1.0
    N = length(background)
    
    prob = hcat([ProbabilityWeights(proba_DNA_gibbs_without_deg(pos_mut, background, h, J, N, T)) for pos_mut in 1:N]...)
    #println(size(prob))
    return get_entropy(prob)[:]
end

function proba_DNA_gibbs_without_deg(k, mutated_seq, h, J,N,  T = 1)
	prob = zeros(20)
	for i in 1:20
        q_k = i
		log_proba = h[q_k, k]
 		for j in 1:N
			log_proba += J[mutated_seq[j], q_k , j, k]
		end
		prob[i] = exp(log_proba/T)
	end
	return normalize(prob,1)
    
end


function cde_1site(site, background, h, J)
    T = 1.0
    N = length(background)
    prob = ProbabilityWeights(proba_DNA_gibbs_without_deg(site, 
            background, h, J, N, T))
    return get_entropy(prob)
end
```

```julia
function find_indices_above_threshold(matrix, threshold, exclude_indices)
    indices_above_threshold = [(i, j) for i in 1:size(matrix, 1), j in 1:size(matrix, 2) if matrix[i, j] > threshold]
    filtered_indices = filter(idx -> idx â exclude_indices, indices_above_threshold)
    return filtered_indices
end


function get_exp_cont(f_exp_rounds, wt, wt_idx, tresh, h, J)
    
    mean_res = []
    previous_muts = []
    i = 1
    n_muts = []
    for f in f_exp_rounds
        if i == 1
            muts = find_indices_above_threshold(f, tresh, wt_idx)
            dE = check_exp_muts_in_wt(muts, wt, h, J)
            push!(mean_res, mean(dE))
            previous_muts = muts
            i += 1
            push!(n_muts, length(muts))
        else
            muts = find_indices_above_threshold(f, tresh, 
                vcat(wt_idx, previous_muts))
            dE = check_exp_muts_in_wt(muts, wt, h, J)
            push!(mean_res, mean(dE))
            previous_muts = muts
            i += 1
            push!(n_muts, length(muts))
        end
    end
    println("Selecting muts with f > $(tresh)")
    print("Number of muts ")
    println(n_muts[:])
    return mean_res
end

function get_exp_cont_with_std(f_exp_rounds, wt, wt_idx, tresh, h, J)
    
    mean_res = []
    std_res = []
    previous_muts = []
    i = 1
    n_muts = []
    for f in f_exp_rounds
        if i == 1
            muts = find_indices_above_threshold(f, tresh, wt_idx)
            dE = check_exp_muts_in_wt(muts, wt, h, J)
            push!(mean_res, mean(dE))
            push!(std_res, std(dE))
            previous_muts = muts
            i += 1
            push!(n_muts, length(muts))
        else
            muts = find_indices_above_threshold(f, tresh, 
                vcat(wt_idx, previous_muts))
            dE = check_exp_muts_in_wt(muts, wt, h, J)
            push!(mean_res, mean(dE))
            push!(std_res, std(dE))
            previous_muts = muts
            i += 1
            push!(n_muts, length(muts))
        end
    end
    return mean_res, std_res
end
```

```julia
function find_common_muts(lists::Vector{Vector{T}}) where T
    if isempty(lists)
        throw(ArgumentError("At least one list must be provided"))
    end
    
    if any(isempty, lists)
        return Int[]
    end
    
    matching_indices = Set(1:length(lists[1]))
    
    for lst in lists[2:end]
        matching_indices = intersect(matching_indices, findall(t -> t in lst, lists[1]))
    end
    
    return collect(matching_indices)
end

function number_of_seqs(z, f_exp, tresh, wt_idx)
    res = []
    for t in tresh

        mut_lists = [find_indices_above_threshold(f, t, wt_idx) 
            for f in f_exp]
        muts_idx = find_common_muts(mut_lists)
        common_muts = mut_lists[1][muts_idx]
        m = []
    
    
        for mut in common_muts 
            for msa in z
                push!(m, sum(msa[mut[2],:] .== mut[1]))
            end
        end
        println(m)
        println((t, minimum(m)))
        push!(res, minimum(m))
    end
    return res
end

function sample_exp_seqs(mut, msa, n_exp_seqs)
    available_seqs = msa[:, msa[mut[2],:] .== mut[1]]
    #println(n_exp_seqs) 
    #println(size(available_seqs, 2))
    if n_exp_seqs > size(available_seqs, 2)
        println("Please compute the mean of deltaE on fewer \n
            experimental sequences, there are not that many available \n
            for the selected mutation")
    end
    col_idx = sample(1:size(available_seqs, 2), n_exp_seqs, replace = false)
    return available_seqs[:, col_idx]
end

function check_1_wt_amino_in_exp_seqs(msa, mut, wt, h, J)
    dEs = zeros(size(msa,2))
    for l in 1:size(msa,2)
        old_seq = deepcopy(msa[:,l])
        new_seq = deepcopy(msa[:,l])
        new_seq[mut[2]] = wt[mut[2]] 
        dEs[l] = delta_energy(h, J, new_seq, old_seq)
    end
    return mean(dEs), std(dEs)
end

function give_exp_entren(f_exp_rounds, z, n_seqs, wt,
        tresh, h, J)
    
    mut_lists = [find_indices_above_threshold(f, tresh, wt_idx) 
            for f in f_exp_rounds]

    
    #println(typeof(mut_lists))
    muts_idx = find_common_muts(mut_lists)
    common_muts = mut_lists[1][muts_idx]
    
    if length(common_muts) .== 0
        println("lower the thresh, no common muts found in exp rounds")
    else
        print("Number of common muts at treshold $(tresh) ")
        println(length(common_muts))
    end
    
    mean_muts =[]
    st_muts = []
    
    for round in z
        
        m_all = []
        st_all = []
        for mut in common_muts
            sub_round = sample_exp_seqs(mut, round, n_seqs)
            m, st = check_1_wt_amino_in_exp_seqs(sub_round, mut, wt, h, J)
            push!(m_all, m)
            push!(st_all, st)
            m_tot = mean(m_all)
            st_tot = (st_all .^2)./(sqrt(length(st_all)))
        end
        m_tot = mean(m_all)
        st_tot = sum(st_all .^2)./(sqrt(length(st_all)))
        push!(mean_muts, m_tot)
        push!(st_muts, st_tot)
    end
    
    #println(st_muts)
    
    return mean_muts
end
            
   
function give_exp_entren_with_std(f_exp_rounds, z, n_seqs, wt,
        tresh, h, J)
    
    mut_lists = [find_indices_above_threshold(f, tresh, wt_idx) 
            for f in f_exp_rounds]

    
    #println(typeof(mut_lists))
    muts_idx = find_common_muts(mut_lists)
    common_muts = mut_lists[1][muts_idx]
    
    if length(common_muts) .== 0
        println("lower the thresh, no common muts found in exp rounds")
    else
        print("Number of common muts at treshold $(tresh) ")
        println(length(common_muts))
    end
    
    mean_muts =[]
    st_muts = []
    
    for round in z
        
        m_all = []
        st_all = []
        for mut in common_muts
            sub_round = sample_exp_seqs(mut, round, n_seqs)
            m, st = check_1_wt_amino_in_exp_seqs(sub_round, mut, wt, h, J)
            push!(m_all, m)
            push!(st_all, st)
            m_tot = mean(m_all)
            st_tot = (st_all .^2)./(sqrt(length(st_all)))
        end
        m_tot = mean(m_all)
        st_tot = sum(st_all .^2)./(sqrt(length(st_all)))
        push!(mean_muts, m_tot)
        push!(st_muts, st_tot)
    end
    
    
    return mean_muts, st_muts
end
```

```julia
function compute_single_ddE_cont(h::Array{Float64,2}, J::Array{Float64,4}, 
        wt::Array{Int8, 1}, seq::Array{Int8, 1}, site::Int64)
    new_wt = deepcopy(wt)
    new_seq = deepcopy(seq)
    new_wt[site] = seq[site]
    new_seq[site] = wt[site]
    return delta_energy(h, J, new_wt, wt) .+ delta_energy(h, J, new_seq, seq)
end

function compute_ddE_cont!(dde, h::Array{Float64,2}, J::Array{Float64,4}, 
        wt::Array{Int8, 1}, seq::Array{Int8, 1}, filt::Array{Tuple{Int64, Int64}, 1})
    
    mut_site = findall(x -> x == 1, wt .!= seq)
    aa_seq = seq[mut_site]
    
    if length(mut_site) == 0
        #println("no muts between the sequences")
    else
        muts = [(aa_seq[i], mut_site[i]) for i in 1:length(aa_seq)]
        filt_muts = filter(t -> t in muts, filt)
        d = Int(KitMSA.my_hamming(wt, seq))
        push!(dde[d], [compute_single_ddE_cont(h, J, wt, seq, mut[2]) 
            for mut in filt_muts]...)
        #println(d)
    end
end

function get_exp_cont2(max_ham, f_exp_rounds, z, wt, wt_idx, tresh, h, J)
    
    mean_res = []
    previous_muts = []
    i = 1
    dde = Vector{Vector{Float64}}(undef, 100)
    for i in 1:100
        dde[i] = Vector{Float64}[]
    end

    for f in f_exp_rounds
        if i == 1
            println("round $i")
            muts = find_indices_above_threshold(f, tresh, wt_idx)
            for n in 1:size(z[i])[2]
                compute_ddE_cont!(dde, h, J, wt, z[i][:,n], muts)
            end
            println((i, length(muts)))
            #push!(previous_muts, muts...)
            i+=1
            println(length(previous_muts))
        else
            println("round $i")
            muts = find_indices_above_threshold(f, tresh, 
                vcat(wt_idx, previous_muts))
            for n in 1:size(z[i])[2]
                compute_ddE_cont!(dde, h, J, wt, z[i][:,n], muts)
            end
            println((i, length(muts)))
            #push!(previous_muts, muts...)
            i+=1
            println(length(previous_muts))
        end
    end
    return dde[1:max_ham]
end           
```
